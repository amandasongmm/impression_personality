```python
def plot_std_value(feat_key):
    x = std_df[feat_key]
    y = final_df[feat_key]
    cor, p = pearsonr(x, y)
    print cor, p
    plt.scatter(x, y)
    plt.xlabel('std')
    plt.ylabel('ave score')
    title_str = ', rho={}, p={}'.format(round(cor, 2), round(p, 4))
    plt.title(feat_key + title_str)
    plt.show()
```
