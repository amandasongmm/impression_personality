```python
resgender = {}
flag=0
with open("mit2k.txt", "r") as infile:
    for line in infile:
        if(flag==0):
            flag=1
            continue
        key = line.split()[0]
        val = line.split()[2]
        if(val=="Male"):
            resgender[key]="1"
        else:
            resgender[key]="0"
```

```python
resgender
```

```python
resage = {}
flag=0
with open("mit2k.txt", "r") as infile:
    for line in infile:
        if(flag==0):
            flag=1
            continue
        key = line.split()[0]
        val = line.split()[3]
        if(val<"20"):
            resage[key]="1"
        elif(val<"30"):
            resage[key]="2"
        elif(val<"45"):
            resage[key]="3"
        elif(val<"60"):
            resage[key]="4"
        else:
            resage[key]="5"
```

```python
resage
```

```python
resethnicity = {}
flag=0
with open("mit2k.txt", "r") as infile:
    for line in infile:
        if(flag==0):
            flag=1
            continue
        key = line.split()[0]
        val = line.split()[1]
        if(val=="WHITE"):
            resethnicity[key]="1"
        elif(val=="BLACK"):
            resethnicity[key]="2"
        else:
            resethnicity[key]="3"
```
